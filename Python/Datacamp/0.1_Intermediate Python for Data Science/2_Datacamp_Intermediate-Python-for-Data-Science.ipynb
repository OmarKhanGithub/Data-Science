```python
# Once you know the distribution, you can start calculating the chances

import matplotlib.pyplot as plt

import numpy as np
np.random.seed(123)
# this list will contain the number of tail you end up with if you played 10 times
final_tails = []
for x in range(10000):
    tails = [0] 
    # start the game of throwing
    for x in range(10):
        coin = np.random.randint(0,2) # flip a coin to get value 1 or 0
        tails.append(tails[x] + coin) # since tails is 1, we can add that to the list element
    # the for loop is one value behind the range so range value of the array already exists
    # and we can call it, and use it in the functions
    # append the final tails value from the game to the upper most list
    final_tails.append(tails[-1])

plt.hist(final_tails, bins = 10)
plt.show()

# -------------------------------------


```

```python
np.random.seed(123)

# Initialize all_walks
all_walks = []

# Simulate random walk 10 times
for i in range(100) :

    # Code from before
    random_walk = [0]
    for x in range(250) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
        
        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0


    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)

```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(123)
all_walks = []

# Simulate random walk 500 times
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1]
print(ends)
# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
# The ends array is everything you need to know the odds; it's available 
# in your Python session so you can make calculations in the IPython Shell.
np.mean(ends >= 60)

# the percentage of beating 60
```
